```python
new = dataset.drop(["customerID","SeniorCitizen"],axis="columns")
```

```python
new['gender'] = new['gender'].astype('category')
new['Partner'] = new['Partner'].astype('category')
new['Dependents'] = new['Dependents'].astype('category')
new['PhoneService'] = new['PhoneService'].astype('category')
new['MultipleLines'] = new['MultipleLines'].astype('category')
new['InternetService'] = new['InternetService'].astype('category')
new['OnlineSecurity'] = new['OnlineSecurity'].astype('category')
new['OnlineBackup'] = new['OnlineBackup'].astype('category')
new['DeviceProtection'] = new['DeviceProtection'].astype('category')
new['TechSupport'] = new['TechSupport'].astype('category')
new['StreamingTV'] = new['StreamingTV'].astype('category')
new['StreamingMovies'] = new['StreamingMovies'].astype('category')
new['Contract'] = new['Contract'].astype('category')
new['PaperlessBilling'] = new['PaperlessBilling'].astype('category')
new['PaymentMethod'] = new['PaymentMethod'].astype('category')
new['TotalCharges'] = pd.to_numeric(new['TotalCharges'], errors='coerce')
```

```python
new.dtypes
```

```python
coree = new[["tenure","MonthlyCharges","TotalCharges"]]
```

```python
corr = coree.corr()
```

```python
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns,annot=True)
```
